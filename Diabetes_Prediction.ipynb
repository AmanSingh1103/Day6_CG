```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

file_name = list(uploaded.keys())[0]
df = pd.read_csv(file_name)

X = df[['BMI']]
y = df['Diabetes']

model = LogisticRegression()
model.fit(X, y)

bmi_range = np.linspace(df['BMI'].min(), df['BMI'].max(), 200).reshape(-1, 1)
probabilities = model.predict_proba(bmi_range)[:, 1]

plt.scatter(df['BMI'], df['Diabetes'])
plt.plot(bmi_range, probabilities)
plt.xlabel("BMI")
plt.ylabel("Diabetes Probability")
plt.title("BMI vs Diabetes (Logistic Regression Curve)")
plt.show()
```
